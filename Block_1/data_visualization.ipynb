```python
import pandas as pd 
import json 
import configparser
from sqlalchemy import create_engine
from sqlalchemy.sql import text

import plotly.express as px
import plotly.graph_objects as go

pd.set_option('display.max_columns', None)
```

```python
config = configparser.ConfigParser()
config.read('.config.ini')

DBHOST = config['DATABASE']['DBHOST']
DBUSER = config['DATABASE']['DBUSER']
DBPASS = config['DATABASE']['DBPASS']
DBNAME = config['DATABASE']['DBNAME']
PORT = config['DATABASE']['PORT']
engine =  create_engine(f"postgresql+psycopg2://{DBUSER}:{DBPASS}@{DBHOST}:{PORT}/{DBNAME}", echo=True)
```

```python
stmt = text("SELECT * FROM booking_data")

# Create a dataFrame using SQL
df = pd.read_sql(stmt, engine)
```

```python
print(f"Number of rows : {df.shape[0]}")
print()

print("Display of dataset: ")
display(df.head(2))
print()

print("Basics statistics: ")
display(df.describe(include='all'))
print()

print("Percentage of missing values: ")
display(100 * df.isnull().sum() / df.shape[0])
```

```python
#Clean up number_of_properties columns
df['number_of_properties'] = df['number_of_properties'].replace('{}', '0')
df['number_of_properties'] = df['number_of_properties'].str.extract('(\d+)')

# Create lat long hotel
df['hotel_lat'], df['hotel_long'], = round(df.hotel_geoposition.apply(lambda x : x.split(",")).str[0].astype(float),2), round(df.hotel_geoposition.apply(lambda x : x.split(",")).str[1].astype(float),2)

# Create column for number of facilities per hotel
df['nb_of_facilities_per_hotel']= df['hotel_popular_facilities'].apply(lambda x : x[1:-1]).apply(lambda x : x.split(",")).apply(lambda x : len(x))
```

```python
# Plot our hotels based on the weather and hotel score

# Sorted dataframe based on max temperature
sorted_df = df.sort_values('temp_max', ascending=True)

fig = px.scatter_mapbox(sorted_df, lat="hotel_lat",
                        lon="hotel_long",
                        mapbox_style="carto-positron",
                        color="hotel_score", size="hotel_starts",
                        color_continuous_scale=px.colors.cyclical.IceFire,
                        animation_frame= 'temp_max',
                        animation_group='city_to_scrap',
                        size_max=15,
                        zoom=14
                        )

fig.update_layout(
    autosize=True,
    paper_bgcolor="LightSteelBlue",
    height=850,
    title=dict(
        text='Map of the best hotel based on hotel score and weather'),
        title_font=dict(size=30)
)


fig.show()
```
